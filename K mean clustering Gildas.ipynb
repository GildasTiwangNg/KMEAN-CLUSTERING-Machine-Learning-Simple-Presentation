```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(df['Age'], df['Income($)'])
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=3)
km
```

```python
y_predicted=km.fit_predict(df[['Age', 'Income($)']])
y_predicted
```

```python
df['cluster']=y_predicted
```

```python
df
```

```python
df0=df[df.cluster==0]
df1=df[df.cluster==1]
df2=df[df.cluster==2]
```

```python
plt.scatter(df0['Age'], df0['Income($)'], marker='*', color='b')
plt.scatter(df1['Age'], df1['Income($)'], marker='+', c='r')
plt.scatter(df2['Age'], df2['Income($)'], c='g')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
df['Income($)']=scaler.fit_transform(df[['Income($)']])
```

```python
df
```

```python
df['Age']=scaler.fit_transform(df[['Age']])
```

```python
df
```

```python
km=KMeans(n_clusters=3)
```

```python
y_predicted=km.fit_predict(df[["Age","Income($)"]])
y_predicted
```

```python
df['cluster2']=y_predicted
```

```python
df.drop(["Income($)", "Name", "cluster"], axis='columns')
```

```python
df0=df[df.cluster2==0]
df1=df[df.cluster2==1]
df2=df[df.cluster2==2]
plt.scatter(df0['Age'], df0['Income_scaled'], marker='*', color='b', label="low")
plt.scatter(df1['Age'], df1['Income_scaled'], marker='+', c='r', label="meduim")
plt.scatter(df2['Age'], df2['Income_scaled'], c='g', label='high')
plt.xlabel('Age')
plt.ylabel('Income_scaled')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color='Black', marker='+', label='centroid')
plt.legend()
```

## Elbow method

```python
k_rng=range(1,10)
sse=[]
for k in k_rng:
    km=KMeans(n_clusters=k)
    km.fit(df[['Age','Income($)']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel("sum of square error")
plt.plot(k_rng, sse)
```
